```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

Data_set=np.loadtxt("/content/sample_data/income_data.csv", delimiter=",")

X=Data_set[:,0]
Y=Data_set[:,1]

# X를 2D 배열로 reshape
X = X.reshape(-1, 1)  # (샘플 수, 특성 수) 형태로 변환

# 모델 생성
model = Sequential()
model.add(Input(shape=(1,)))  # 입력이 1개 특성임을 정의
model.add(Dense(30, activation='linear'))  # 30개의 유닛을 가진 밀집 층 추가
model.add(Dense(1, activation='linear'))  # 최종 출력층 (1개의 유닛)

# 모델 컴파일
model.compile(loss='mse', optimizer='sgd')  # 손실 함수는 평균 제곱 오차(mse), 옵티마이저는 SGD

# 모델 학습
model.fit(X, Y, epochs=100, batch_size=10)  # 모델 학습
```

```python
import matplotlib.pyplot as plt
predictions=model.predict(X)

plt.scatter(Y, predictions, color='blue', label='Predicted vs Actual')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vd Predicted Values for Linear Regression')

plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'r--', lw=2)
plt.legend()
plt.show()
```
